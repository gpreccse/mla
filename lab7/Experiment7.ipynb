```python
#data loading
import pandas as pd
data = pd.read_csv('iris.csv')
data.head()
```

```python
#data extraction
x = data.drop('target',axis=True)
y = data[['target']]
```

```python
#data splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
#First type Normal
#model selection
from sklearn.svm import SVC
model = SVC()
#model training
model.fit(x_train,y_train)
#model prediction
y_pred = model.predict(x_test)
#model performance
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#Second Type with C value
#model selection
model = SVC(C=0.1)
#model training
model.fit(x_train,y_train)
#model prediction
y_pred = model.predict(x_test)
#model performance
accuracy_score(y_test,y_pred)
```

```python
#Third Type with gamma value
#model selection
model = SVC(gamma=0.1)
#model training
model.fit(x_train,y_train)
#model prediction
y_pred = model.predict(x_test)
#model performance
accuracy_score(y_test,y_pred)
```

```python
#Fourth Type with C,gamma,kernal value
#model selection
model = SVC(C=0.1,gamma=0.1,kernel='linear')
#model training
model.fit(x_train,y_train)
#model prediction
y_pred = model.predict(x_test)
#model performance
accuracy_score(y_test,y_pred)
```
